1D Array Manipulation: Create a 1D NumPy array with 50 random integers ranging from 0 to 100. Perform the following operations:
*Find the sum of all elements. *Identify the maximum and minimum values. *Calculate the mean and standard deviation.

```python
import numpy as np
arr1 = np.arange(0,100)
add = np.sum(arr1)
maxi = np.max(arr1)
mini = np.min(arr1)
mean = np.mean(arr1)
std = np.std(arr1)
print(f"\n The sum of all elements is {add}",f"\n The maximum and minimum values are {maxi} and {mini}", f"\n The mean and standard deviation are {mean} and {std:.2f}")
```

2.2D Array Operations: Create a 5x5 matrix with random numbers between 1 and 10.Replace all odd numbers in the matrix with -1. *Find the sum of all elements in each row and each column. *Compute the transpose of the matrix.

```python
import numpy as np
arr2 = np.random.randint(1, 11, size=(5, 5))
print("Original Matrix:")
print(arr2)

arr2[arr2 % 2 != 0] = -1
print("\nMatrix after replacing odd numbers with -1:")
print(arr2)

row_sums = np.sum(arr2, axis=1)
col_sums = np.sum(arr2, axis=0)
print(f"\nSum of each row: {row_sums}")
print(f"Sum of each column: {col_sums}")

transpos = np.transpose(arr2)
print("\nTranspose of the matrix:")
print(transpos)
```

4.Image Manipulation: Load an image using NumPy and perform the following operations:Convert the image to grayscale.Apply a Gaussian blur filter using NumPy functions.Rotate the image by 90 degrees clockwise.


```python
import numpy as np
from skimage import data, color
from scipy.ndimage import gaussian_filter
import matplotlib.pyplot as plt

image = data.astronaut()  
print("Original Image Shape:", image.shape)

gray_image = color.rgb2gray(image)
print("Grayscale Image Shape:", gray_image.shape)

blurred_image = gaussian_filter(gray_image, sigma=2)
print("Blurred Image Shape:", blurred_image.shape)

rotated_image = np.rot90(blurred_image, k=-1)  # k=-1 means 90 degrees clockwise
print("Rotated Image Shape:", rotated_image.shape)

# Plot the subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(gray_image, cmap='gray')
axes[0].set_title("Grayscale Image")

axes[1].imshow(blurred_image, cmap='gray')
axes[1].set_title("Blurred Image")

axes[2].imshow(rotated_image, cmap='gray')
axes[2].set_title("Rotated Image (90 Clockwise)")

plt.show()
```

5. Statistical and Universal Functions: Create a 3x4 matrix with random floats between 0 and 1. Perform the following operations:Apply the exp universal function to the matrix.Calculate the sum, mean, and variance of each column.Normalize the matrix using min-max scaling.

```python
import numpy as np

arr4 = np.random.rand(3, 4)
print(f"Original Matrix:\n",arr4)

exp_matrix = np.exp(arr4)
print(f"\nMatrix after applying the exp function: \n",exp_matrix)

col_sum = np.sum(arr4, axis=0)
col_mean = np.mean(arr4, axis=0)
col_variance = np.var(arr4, axis=0)

print(f"\nSum of each column\n: {col_sum}")
print(f"Mean of each column\n: {col_mean}")
print(f"Variance of each column\n: {col_variance}")

min_val = np.min(arr4, axis=0)
max_val = np.max(arr4, axis=0)
normalized_matrix = (arr4 - min_val) / (max_val - min_val)

print("\nNormalized Matrix (Min-Max Scaling):")
print(normalized_matrix)
```
