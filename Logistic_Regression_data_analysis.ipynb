```python
sex=pd.get_dummies(titanic['sex'],drop_first=True)
sex.head(5)
```

```python
embark=pd.get_dummies(titanic["embarked"])
embark.head(5)
```

```python
embark=pd.get_dummies(titanic["embarked"],drop_first=True)
embark.head(5)
```

```python
pcl=pd.get_dummies(titanic["pclass"],drop_first=True)
pcl.head(5)
```

```python
titanic=pd.concat([titanic,sex,embark,pcl],axis=1)
```

```python
titanic.head(5)
```

```python
titanic.drop(['sex','class','adult_male','embark_town'],axis=1,inplace=True)
```

```python
titanic.drop(['who','alive','embarked'],axis=1,inplace=True)
```

```python
titanic.drop(['alone'],axis=1,inplace=True)
```

```python
titanic.head(5)
```

```python
# training data
```

```python
X = titanic.drop("survived",axis=1)
y = titanic["survived"]
```

```python
X.columns = X.columns.astype(str)
```

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,predictions)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,predictions)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predictions)
```
