1. Create a DataFrame with at least 4 columns and 5 rows. Write a script to add a new column and then remove an existing column from the DataFrame.


```python
import pandas as pd

data = {
    'A': [1, 2, 3, 4, 5],
    'B': ['a', 'b', 'c', 'd', 'e'],
    'C': [10.5, 11.5, 12.5, 13.5, 14.5],
    'D': [True, False, True, False, True]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

df['E'] = [100, 200, 300, 400, 500]
print("\nDataFrame after adding a new column 'E':")
print(df)

df.drop(columns=['B'], inplace=True)
print("\nDataFrame after removing the column 'B':")
print(df)
```

2. Create a DataFrame and group the data based on a particular column, then calculate the mean for each group.


```python
import pandas as pd

data = {
    'Category': ['A', 'B', 'A', 'B', 'C', 'C', 'A', 'B'],
    'Value1': [10, 20, 30, 40, 50, 60, 70, 80],
    'Value2': [5, 10, 15, 20, 25, 30, 35, 40]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

grouped_df = df.groupby('Category')

mean_values = grouped_df.mean()
print("\nMean values for each group:")
print(mean_values)
```

3. Write a function to drop specific rows based on a condition from a DataFrame.


```python
import pandas as pd

def drop_rows_based_on_condition(df, condition):
    filtered_df = df[~condition(df)]
    return filtered_df

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [24, 30, 22, 35, 29],
    'Salary': [70000, 80000, 120000, 60000, 95000]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

condition = lambda df: df['Age'] > 30

new_df = drop_rows_based_on_condition(df, condition)

print("\nDataFrame after dropping rows where Age > 30:")
print(new_df)
```

Matplotlib Question:

4. Create a figure with specific dimensions (e.g., 8x6), plot multiple arrays, and use different line colors and markers for each array. Set appropriate labels and a title.


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)  
y1 = np.sin(x)                
y2 = np.cos(x)                
y3 = np.sin(x) + np.cos(x)   

plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='Sine', color='blue', marker='o', linestyle='-')
plt.plot(x, y2, label='Cosine', color='orange', marker='s', linestyle='--')
plt.plot(x, y3, label='Sine + Cosine', color='green', marker='^', linestyle=':')
plt.title('Sine and Cosine Functions')
plt.xlabel('X-axis (radians)')
plt.ylabel('Y-axis (values)')
plt.legend()  
plt.grid()
plt.tight_layout()  
plt.show()
```

5. Using Seaborn, visualize the distribution of data in a dataset (you can either take own data or  download a dataset from Kaggle). Create a bar plot to show the relationship between two categorical variables.


```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')

plt.figure(figsize=(8, 5))
sns.histplot(titanic['age'], bins=30, kde=True, color='blue')
plt.title('Distribution of Ages on the Titanic')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid()
plt.show()
plt.figure(figsize=(8, 5))
sns.countplot(data=titanic, x='class', hue='sex', palette='Set2')
plt.title('Count of Passengers by Class and Sex')
plt.xlabel('Class')
plt.ylabel('Count')
plt.legend(title='Sex')
plt.grid(axis='y')
plt.show()
```


6. Using Seaborn, create a pair plot to analyze the relationships between numerical variables in a dataset,.(take iris dataset from seaborn builtin dataset).

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

plt.figure(figsize=(10, 8))
pair_plot = sns.pairplot(iris, hue='species', markers=['o', 's', 'D'], palette='Set2')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02) 
plt.show()
```
